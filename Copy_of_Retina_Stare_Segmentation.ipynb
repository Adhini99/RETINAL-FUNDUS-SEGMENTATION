```python
import os
os.environ["TF_FORCE_GPU_ALLOW_GROWTH"] = "true"
import numpy as np
import cv2
from PIL import Image
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Dropout, Flatten, Dense, BatchNormalization, ReLU
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import roc_auc_score, accuracy_score, confusion_matrix, f1_score, jaccard_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import glob
from IPython.display import Image as IPyImage, display
import base64
from io import BytesIO

# Ensure matplotlib displays in notebook
%matplotlib inline

# Corrected STARE dataset paths based on your input
TRAIN_IMAGES_PATH = "/kaggle/input/stare-dataset-20images/stare-images"
TRAIN_MANUAL_PATH = "/kaggle/input/stare-dataset-20images/labels-ah" # Using Hoover's annotations

# For testing, we'll assume a similar structure or re-use parts of the dataset
TEST_IMAGES_PATH = "/kaggle/input/stare-dataset-20images/stare-images"
TEST_MANUAL_PATH = "/kaggle/input/stare-dataset-20images/labels-ah"

def check_directory(path, description="directory"):
    """Checks if a directory exists and prints its contents (first 5 files) and total count."""
    if os.path.exists(path):
        all_files = glob.glob(os.path.join(path, "*"))
        # Print a sample of files and the total count
        print(f"Contents of {description} ({path}): {[os.path.basename(f) for f in all_files[:5]]} ... ({len(all_files)} files total)")
    else:
        print(f"{description} ({path}) does not exist!")
    return os.path.exists(path)

def load_image(file_path):
    """Loads an image, extracting the green channel for RGB images or converting to grayscale for others."""
    try:
        img = Image.open(file_path)
        if img.mode == "RGB":
            img = np.array(img)
            green_channel = img[:, :, 1]  # STARE images are RGB, green channel often best for vessels
            return green_channel
        # For manual annotations, they might be grayscale or indexed. Convert to grayscale.
        return np.array(img.convert('L'))
    except Exception as e:
        print(f"Error loading image {file_path}: {e}")
        return None

def generate_fov_mask(image_shape):
    """Generates a circular Field Of View (FOV) mask for the given image shape."""
    height, width = image_shape
    mask = np.zeros((height, width), dtype=np.uint8)
    center = (width // 2, height // 2)
    # Radius adjusted for STARE's typical image sizes, with a small buffer
    radius = min(height, width) // 2 - 30
    y, x = np.ogrid[:height, :width]
    distance = np.sqrt((x - center[0])**2 + (y - center[1])**2)
    mask[distance <= radius] = 255
    return mask

def preprocess_image(image):
    """Applies Global Contrast Normalization (GCN), CLAHE, and Gamma Correction to an image."""
    if image is None:
        return None
    image_float = image.astype(np.float32)

    # Global Contrast Normalization
    mean = np.mean(image_float)
    std = np.std(image_float) if np.std(image_float) != 0 else 1
    gcn_image = (image_float - mean) / std
    gcn_image = np.clip(gcn_image, -3, 3) # Clip outliers
    gcn_image = ((gcn_image + 3) / 6 * 255).astype(np.uint8) # Scale back to 0-255

    # CLAHE (Contrast Limited Adaptive Histogram Equalization)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    clahe_image = clahe.apply(gcn_image)

    # Gamma Correction
    gamma = 1.2 # Common value for retinal images
    gamma_image = np.array(255 * (clahe_image / 255.0) ** (1.0 / gamma), dtype=np.uint8)
    return gamma_image

def extract_patches(image, mask, manual=None, patch_size=64, max_patches=5000):
    """
    Extracts overlapping patches from an image within the FOV mask.
    Optionally extracts corresponding labels from a manual annotation.
    """
    if image is None or mask is None or (manual is not None and manual is None):
        print("Invalid input image, mask, or manual. Returning empty patches.")
        return np.array([]), np.array([]) if manual is not None else None, []

    patches = []
    labels = []
    coords = []
    height, width = image.shape
    step = patch_size // 4 # Overlapping patches
    pad_size = patch_size // 2 # Padding to ensure patches can be extracted from edges

    # Pad images and masks to extract patches from boundaries
    padded_image = np.pad(image, ((pad_size, pad_size), (pad_size, pad_size)), mode='constant')
    padded_mask = np.pad(mask, ((pad_size, pad_size), (pad_size, pad_size)), mode='constant')

    padded_manual = None
    if manual is not None:
        padded_manual = np.pad(manual, ((pad_size, pad_size), (pad_size, pad_size)), mode='constant')

    indices = []
    # Iterate through potential patch centers
    for i in range(pad_size, height + pad_size, step):
        for j in range(pad_size, width + pad_size, step):
            # Only consider patch centers within the FOV mask
            if padded_mask[i, j] == 0:
                continue
            indices.append((i, j))

    # Randomly sample if too many patches
    if max_patches and len(indices) > max_patches:
        np.random.shuffle(indices)
        indices = indices[:max_patches]

    for i, j in indices:
        # Extract patch
        patch = padded_image[i - pad_size : i + pad_size, j - pad_size : j + pad_size]
        if patch.shape != (patch_size, patch_size): # Ensure patch is full size
            continue

        if manual is not None:
            # Label is the central pixel of the manual annotation patch
            label = padded_manual[i, j] // 255 # Convert 0/255 to 0/1
            labels.append(label)
        patches.append(patch)
        coords.append((i - pad_size, j - pad_size)) # Store original (unpadded) top-left coord

    patches = np.array(patches)
    labels = np.array(labels) if manual is not None else None

    if manual is not None:
        if labels.size > 0:
            print(f"Extracted {len(patches)} patches: {np.sum(labels == 1)} vessel, {np.sum(labels == 0)} non-vessel")
        else:
            print(f"Extracted {len(patches)} patches, but no labels found (labels array is empty).")
    else:
        print(f"Extracted {len(patches)} patches for prediction")
    return patches, labels, coords

def level_i_balancing(patches, labels, rho=0.8):
    """
    Performs Level I balancing to address class imbalance between vessel and non-vessel patches.
    It balances vessel patches with two types of non-vessel patches:
    - Partial background (higher mean intensity, often near vessels)
    - Full background (lower mean intensity)
    """
    if labels is None or len(labels) == 0 or len(patches) == 0:
        print("Warning: No valid patches or labels for balancing. Returning empty.")
        return np.array([]), np.array([])

    # Calculate mean intensity for each patch (normalized to 0-1)
    if patches.ndim == 4: # If patches are (N, H, W, 1)
        patch_means = np.mean(patches, axis=(1, 2, 3)) / 255.0
    else: # If patches are (N, H, W)
        patch_means = np.mean(patches, axis=(1, 2)) / 255.0

    vessel_patches = patches[labels == 1]
    non_vessel_patches = patches[labels == 0]
    non_vessel_means = patch_means[labels == 0]

    num_vessel = len(vessel_patches)
    if num_vessel == 0:
        print("Warning: No vessel patches found for balancing. Returning original patches.")
        # If there are no vessels, we should still return the non-vessel patches
        # or handle this case as an empty set if we only care about vessel-related training.
        # For a full segmentation model, we need to learn non-vessels too.
        # For simplicity, if no vessels, return just the non-vessel patches for consistency,
        # but the purpose of balancing is to get more vessels.
        # It's better to just return an empty set if we can't balance properly without vessels.
        return np.array([]), np.array([])

    # Define a threshold to differentiate partial and full background
    # 75th percentile of non-vessel patch means is a common heuristic
    threshold = np.percentile(non_vessel_means, 75) if len(non_vessel_means) > 0 else 0.5

    # Separate non-vessel patches into partial and full background
    partial_background_indices = np.where(non_vessel_means >= threshold)[0]
    full_background_indices = np.where(non_vessel_means < threshold)[0]

    partial_background = non_vessel_patches[partial_background_indices]
    full_background = non_vessel_patches[full_background_indices]

    # Calculate desired number of patches for each background type based on rho
    num_partial_to_select = int(num_vessel * rho)
    num_full_to_select = int(num_vessel * (1 - rho))

    selected_partial = np.array([])
    if len(partial_background) > 0:
        selected_partial_indices = np.random.choice(len(partial_background), min(num_partial_to_select, len(partial_background)), replace=False)
        selected_partial = partial_background[selected_partial_indices]

    selected_full = np.array([])
    if len(full_background) > 0:
        selected_full_indices = np.random.choice(len(full_background), min(num_full_to_select, len(full_background)), replace=False)
        selected_full = full_background[selected_full_indices]

    # Concatenate all selected patches and their labels
    balanced_patches_list = [vessel_patches]
    balanced_labels_list = [np.ones(len(vessel_patches), dtype=int)] # Labels for vessel patches are 1

    if len(selected_partial) > 0:
        balanced_patches_list.append(selected_partial)
        balanced_labels_list.append(np.zeros(len(selected_partial), dtype=int)) # Labels for background are 0
    if len(selected_full) > 0:
        balanced_patches_list.append(selected_full)
        balanced_labels_list.append(np.zeros(len(selected_full), dtype=int))

    balanced_patches = np.concatenate(balanced_patches_list, axis=0)
    balanced_labels = np.concatenate(balanced_labels_list, axis=0)

    print(f"Balanced dataset: {len(vessel_patches)} vessel, {len(selected_partial)} partial background, {len(selected_full)} full background. Total: {len(balanced_patches)} patches")
    return balanced_patches, balanced_labels


def build_model(input_shape=(64, 64, 1)):
    """
    Builds a Convolutional Neural Network (CNN) model for vessel segmentation.
    The architecture is inspired by common CNN designs for image classification/segmentation.
    """
    model = Sequential([
        Input(shape=input_shape),
        Conv2D(32, (3, 3), padding='same'),
        BatchNormalization(),
        ReLU(),
        Conv2D(32, (3, 3), padding='same'),
        BatchNormalization(),
        ReLU(),
        MaxPooling2D((2, 2)),
        Dropout(0.25),

        Conv2D(64, (3, 3), padding='same'),
        BatchNormalization(),
        ReLU(),
        Conv2D(64, (3, 3), padding='same'),
        BatchNormalization(),
        ReLU(),
        MaxPooling2D((2, 2)),
        Dropout(0.25),

        Conv2D(128, (3, 3), padding='same'),
        BatchNormalization(),
        ReLU(),
        Conv2D(128, (3, 3), padding='same'),
        BatchNormalization(),
        ReLU(),
        MaxPooling2D((2, 2)),
        Dropout(0.35),

        Flatten(),
        Dense(512),
        BatchNormalization(),
        ReLU(),
        Dropout(0.5),
        Dense(1, activation='sigmoid') # Sigmoid for binary classification (vessel/non-vessel)
    ])
    model.compile(optimizer=Adam(learning_rate=0.0002), loss='binary_crossentropy', metrics=['accuracy'])
    return model

def postprocess_mask(pred_mask, original_mask):
    """
    Applies morphological operations (closing and opening) and then applies the FOV mask
    to refine the predicted vessel segmentation.
    """
    pred_mask_binary = (pred_mask > 127).astype(np.uint8) # Convert to binary 0/1 for morphological ops

    # Define a small kernel for morphological operations
    # The size of the kernel can be tuned. Larger kernels affect larger features.
    kernel_size = 3 # A smaller kernel might preserve finer details
    kernel = np.ones((kernel_size, kernel_size), np.uint8)

    # Closing: Fills small holes and connects nearby features
    # Iterations can be tuned, typically 1 or 2
    closed_mask = cv2.morphologyEx(pred_mask_binary, cv2.MORPH_CLOSE, kernel, iterations=1)

    # Opening: Removes small objects (noise) and smooths contours
    opened_mask = cv2.morphologyEx(closed_mask, cv2.MORPH_OPEN, kernel, iterations=1)

    # Apply the original FOV mask to ensure predictions are only within the valid retinal area
    original_mask_binary = (original_mask > 0).astype(np.uint8) # Ensure FOV mask is binary 0/1

    # Result should be 0 or 255 for visualization
    final_mask = opened_mask * original_mask_binary * 255
    return final_mask.astype(np.uint8)

def predict_mask(model, image, mask, patch_size=64, threshold=0.5):
    """
    Generates a full-image vessel segmentation mask using the trained patch-based model.
    Averages probabilities from overlapping patches, then applies threshold.
    """
    if image is None or mask is None:
        print("Invalid image or mask for prediction.")
        return np.zeros_like(image, dtype=np.uint8)

    # Preprocess the original image
    preprocessed_image = preprocess_image(image)
    if preprocessed_image is None:
        print("Preprocessing failed for prediction.")
        return np.zeros_like(image, dtype=np.uint8)

    # Apply FOV mask to the preprocessed image
    preprocessed_image_masked = preprocessed_image * (mask // 255) # Ensure mask is 0/1 for multiplication

    # Extract patches for prediction (no labels needed here)
    # max_patches=None ensures all possible patches are extracted for full image reconstruction
    patches, _, coords = extract_patches(preprocessed_image_masked, mask, manual=None, patch_size=patch_size, max_patches=None)
    if len(patches) == 0:
        print("No patches extracted for prediction. Returning empty mask.")
        return np.zeros_like(image, dtype=np.uint8)

    # Reshape patches for model input (add channel dimension)
    patches = patches.reshape(-1, patch_size, patch_size, 1)

    # Get probability predictions from the model (values between 0.0 and 1.0)
    pred_probabilities = model.predict(patches, verbose=0).flatten()

    height, width = image.shape
    # Initialize accumulators for averaging predictions over overlapping regions
    prob_mask_accumulator = np.zeros((height, width), dtype=np.float32)
    patch_count_accumulator = np.zeros((height, width), dtype=np.float32)

    # Reconstruct the full mask by placing predicted probabilities back into their original locations
    for idx, (top, left) in enumerate(coords):
        top_end = min(top + patch_size, height)
        left_end = min(left + patch_size, width)

        # Add the probability of the central pixel of the patch to the accumulator
        # We are assigning the probability of the patch to the entire patch area for simplicity,
        # or more precisely, for the central pixel if that's how your `extract_patches` was designed.
        # Given `extract_patches` takes central pixel for label, let's refine this to
        # only affect the central pixel of the predicted patch or the whole patch with probability.
        # For full mask reconstruction, we usually want to average probabilities over overlaps.

        # A simple way for a patch-based classification model is to assign the probability
        # to the central pixel of where the patch came from. Or, if the model predicts the whole patch,
        # then sum it up over the patch area. Given this is a binary classifier on patches,
        # assigning the probability to the central pixel it classified is more aligned with the labels.
        # But for smooth reconstruction, summing the probabilities over the patch area is common.

        # Let's average the probabilities over the entire patch region
        prob_mask_accumulator[top:top_end, left:left_end] += pred_probabilities[idx]
        patch_count_accumulator[top:top_end, left:left_end] += 1

    # Average the accumulated probabilities for overlapping regions
    epsilon = 1e-6 # To prevent division by zero
    pred_mask_averaged_probs = np.where(patch_count_accumulator > 0,
                                        prob_mask_accumulator / (patch_count_accumulator + epsilon), 0)

    # Apply the optimal threshold to the averaged probabilities to get the binary mask
    # Then scale to 0 or 255 for visualization
    final_pred_mask = (pred_mask_averaged_probs > threshold).astype(np.uint8) * 255

    # Apply FOV mask to the final result (this is crucial to remove predictions outside the eye)
    final_pred_mask = final_pred_mask * (mask // 255) # Mask is 0/255, convert to 0/1 for multiplication

    print(f"Processed {len(patches)} patches for prediction, Predicted mask (before post-processing): {np.sum(final_pred_mask > 0)} vessel pixels")
    return postprocess_mask(final_pred_mask, mask) # Apply post-processing (morphological ops + final FOV mask)


def evaluate_model(model, val_patches, val_labels):
    """
    Evaluates the model on validation patches and determines the optimal threshold
    based on a combination of AUC and F1-score.
    """
    if val_patches is None or val_labels is None or len(val_patches) == 0:
        print("No validation data available for evaluation.")
        return None, None, None, None, None, None, None

    predictions = model.predict(val_patches, verbose=0).flatten() # These are probabilities
    val_labels = val_labels.astype(np.uint8) # Ensure labels are uint8 for scikit-learn

    thresholds = np.linspace(0.3, 0.7, 9) # Test thresholds from 0.3 to 0.7
    best_acc, best_auc, best_sensitivity, best_specificity, best_f1, best_iou = 0, 0, 0, 0, 0, 0
    best_threshold = 0.5 # Default best threshold

    print("\n--- Evaluation Metrics per Threshold ---")
    for threshold in thresholds:
        pred_binary = (predictions > threshold).astype(np.uint8)

        # Calculate confusion matrix components
        tn, fp, fn, tp = 0, 0, 0, 0
        try:
            # handle cases where one class might be missing in actual or predicted
            # This handles cases where val_labels or pred_binary might be all 0s or all 1s
            unique_true = np.unique(val_labels)
            unique_pred = np.unique(pred_binary)

            if len(unique_true) < 2 or len(unique_pred) < 2:
                # If only one class is present in true or pred, confusion_matrix might fail
                # or give a reduced matrix. Handle manually or assign default zeros/ones.
                if np.all(val_labels == 0) and np.all(pred_binary == 0):
                    tn = len(val_labels)
                elif np.all(val_labels == 1) and np.all(pred_binary == 1):
                    tp = len(val_labels)
                elif np.all(val_labels == 0) and np.all(pred_binary == 1):
                    fp = len(val_labels)
                elif np.all(val_labels == 1) and np.all(pred_binary == 0):
                    fn = len(val_labels)
                else: # mixed but one is single class
                     cm = confusion_matrix(val_labels, pred_binary, labels=[0, 1]).ravel()
                     if len(cm) == 4:
                        tn, fp, fn, tp = cm
                     else: # e.g., if only one class predicted, cm will be (1,2) or (2,1)
                         # This case should be rare with proper balancing
                         print(f"Warning: Unexpected confusion matrix shape for threshold {threshold:.2f}, cm: {cm}. Setting to 0.")
                         tn, fp, fn, tp = 0,0,0,0
            else: # Both classes present in true and predicted, normal case
                cm = confusion_matrix(val_labels, pred_binary, labels=[0, 1]).ravel()
                tn, fp, fn, tp = cm
        except ValueError as e:
            print(f"Error calculating confusion matrix for threshold {threshold:.2f}: {e}.")
            tn, fp, fn, tp = 0, 0, 0, 0 # Fallback

        # Calculate metrics
        acc = accuracy_score(val_labels, pred_binary)

        # AUC, F1, IoU require at least two unique true and predicted classes
        # AUC specific check: requires probabilities for `roc_auc_score` not binary
        # Here, `pred_binary` is used for `roc_auc_score` which is a common simplification
        # but technically for optimal AUC, one uses raw probabilities.
        # However, for threshold optimization, this is acceptable.

        # Check unique classes for roc_auc_score
        if len(np.unique(val_labels)) > 1 and len(np.unique(predictions)) > 1:
            auc = roc_auc_score(val_labels, predictions) # Use raw probabilities for AUC
        else:
            auc = 0.5 # Default if AUC cannot be computed

        sensitivity = tp / (tp + fn) if (tp + fn) > 0 else 0
        specificity = tn / (tn + fp) if (tn + fp) > 0 else 0

        if len(np.unique(val_labels)) > 1 and len(np.unique(pred_binary)) > 1:
            f1 = f1_score(val_labels, pred_binary)
            iou = jaccard_score(val_labels, pred_binary)
        else:
            f1 = 0
            iou = 0

        print(f"Threshold {threshold:.2f}: Acc: {acc:.4f}, AUC: {auc:.4f}, Sens: {sensitivity:.4f}, Spec: {specificity:.4f}, F1: {f1:.4f}, IoU: {iou:.4f}")

        # Determine best threshold based on (AUC + F1) / 2
        # Use AUC and F1 from current threshold
        current_metric = (auc + f1) / 2
        if current_metric > (best_auc + best_f1) / 2:
            best_acc, best_auc, best_sensitivity, best_specificity = acc, auc, sensitivity, specificity
            best_f1, best_iou = f1, iou
            best_threshold = threshold
        elif current_metric == (best_auc + best_f1) / 2: # If metrics are equal, prefer lower threshold
            if threshold < best_threshold:
                best_acc, best_auc, best_sensitivity, best_specificity = acc, auc, sensitivity, specificity
                best_f1, best_iou = f1, iou
                best_threshold = threshold

    print(f"\nOptimal Threshold: {best_threshold:.2f}")
    # Using the original 'predictions' (probabilities) to determine distribution
    print(f"Prediction distribution at optimal threshold: {np.sum(predictions > best_threshold)} vessel, {np.sum(predictions <= best_threshold)} non-vessel")
    return best_acc, best_auc, best_sensitivity, best_specificity, best_f1, best_iou, best_threshold

def load_dataset(images_path, manual_path, num_images_to_use=None):
    """
    Loads and pairs original images with their manual annotations from specified directories.
    Handles the specific STARE naming convention (imXXXX.ppm and imXXXX.ah.ppm).
    Extracts and balances patches.
    """
    if not check_directory(images_path, "Image directory") or \
       not check_directory(manual_path, "Manual annotation directory"):
        print("One or both dataset paths do not exist. Exiting.")
        return None, None

    image_files = sorted(glob.glob(os.path.join(images_path, "*.ppm")))
    manual_files = sorted(glob.glob(os.path.join(manual_path, "*.ppm")))

    paired_files = []

    # Corrected logic for extracting base names for matching
    # For original images: im0001.ppm -> 'im0001'
    image_basenames_map = {os.path.basename(f).replace('.ppm', ''): f for f in image_files}

    # For manual annotations: im0001.ah.ppm -> 'im0001'
    manual_basenames_map = {}
    for f in manual_files:
        filename = os.path.basename(f)
        # We specifically look for the Hoover annotation suffix '.ah.ppm'
        if '.ah.ppm' in filename:
            base_name_for_matching = filename.replace('.ah.ppm', '')
            manual_basenames_map[base_name_for_matching] = f
        # If there were other annotation types (e.g., .vk.ppm), you'd add:
        # elif '.vk.ppm' in filename:
        #     base_name_for_matching = filename.replace('.vk.ppm', '')
        #     manual_basenames_map[base_name_for_matching] = f

    # Iterate through image basenames and find corresponding manual basenames
    for basename in sorted(image_basenames_map.keys()):
        if basename in manual_basenames_map:
            paired_files.append((image_basenames_map[basename], manual_basenames_map[basename]))
        else:
            print(f"Warning: No matching Hoover annotation found for image {basename}.ppm. Skipping this image.")

    if not paired_files:
        print("No paired images and manual annotations found. Check file naming and folder structure.")
        return np.array([]), np.array([])

    print(f"Found {len(paired_files)} paired image-manual sets in total.")

    # Limit to num_images_to_use if specified, otherwise use all found paired files
    if num_images_to_use is not None:
        paired_files_for_processing = paired_files[:num_images_to_use]
        print(f"Processing first {len(paired_files_for_processing)} paired image-manual sets.")
    else:
        paired_files_for_processing = paired_files
        print(f"Processing all {len(paired_files_for_processing)} paired image-manual sets.")


    all_patches = []
    all_labels = []

    for idx, (img_file, manual_file) in enumerate(paired_files_for_processing):
        print(f"Processing image {idx + 1}/{len(paired_files_for_processing)}: {os.path.basename(img_file)}")

        image = load_image(img_file)
        manual = load_image(manual_file)

        if image is None:
            print(f"Skipping image {img_file} due to loading error or invalid image data.")
            continue
        if manual is None:
            print(f"Skipping manual {manual_file} due to loading error or invalid image data.")
            continue

        # Ensure manual annotation is binary (0 or 255)
        manual = (manual > 127).astype(np.uint8) * 255

        mask = generate_fov_mask(image.shape)
        preprocessed_image = preprocess_image(image)

        if preprocessed_image is None:
            print(f"Skipping image {os.path.basename(img_file)} due to preprocessing error.")
            continue

        # Apply FOV mask to the preprocessed image before patch extraction
        preprocessed_image_masked = preprocessed_image * (mask // 255) # Mask is 0/255, convert to 0/1 for multiplication

        patches, labels, _ = extract_patches(preprocessed_image_masked, mask, manual)

        if len(patches) == 0 or labels is None or len(labels) == 0:
            print(f"No valid patches or labels extracted for image {os.path.basename(img_file)}. Skipping.")
            continue

        # Perform Level I balancing
        balanced_patches, balanced_labels = level_i_balancing(patches, labels)

        if len(balanced_patches) > 0 and len(balanced_labels) > 0:
            all_patches.append(balanced_patches)
            all_labels.append(balanced_labels)
        else:
            print(f"Balancing resulted in empty patches/labels for {os.path.basename(img_file)}. Skipping.")

    if not all_patches:
        print("No patches loaded from any image after processing and balancing.")
        return np.array([]), np.array([])

    all_patches_concatenated = np.concatenate(all_patches, axis=0)
    all_labels_concatenated = np.concatenate(all_labels, axis=0)

    # Reshape patches for CNN input (add channel dimension)
    all_patches_concatenated = all_patches_concatenated.reshape(-1, 64, 64, 1)

    print(f"Total dataset for training/validation: {len(all_patches_concatenated)} patches, {np.sum(all_labels_concatenated == 1)} vessel, {np.sum(all_labels_concatenated == 0)} non-vessel")
    return all_patches_concatenated, all_labels_concatenated


def main():
    print("Checking dataset directories:")
    check_directory(TRAIN_IMAGES_PATH, "Train Images")
    check_directory(TRAIN_MANUAL_PATH, "Train Manuals")
    # For STARE 20-images, test path is same as train path for loading
    check_directory(TEST_IMAGES_PATH, "Test Images")
    check_directory(TEST_MANUAL_PATH, "Test Manuals")

    # Load all available images and labels from the specified paths
    # The 'stare-dataset-20images' typically implies there are only 20 relevant images,
    # so we'll load all of them and then split for train/val/test.
    all_images_patches, all_images_labels = load_dataset(TRAIN_IMAGES_PATH, TRAIN_MANUAL_PATH, num_images_to_use=None) # Load all

    if all_images_patches is None or len(all_images_patches) == 0:
        print("No data could be loaded. Exiting.")
        return

    # Split the loaded data into training (70%), validation (15%), and testing (15%)
    # This is a common practice when a dataset doesn't have predefined splits.
    # First, split into train and temp (val + test)
    train_patches, temp_patches, train_labels, temp_labels = train_test_split(
        all_images_patches, all_images_labels, test_size=0.3, random_state=42, stratify=all_images_labels
    )

    # Then split temp into validation and test (0.5 of temp_size = 0.15 of original_size)
    val_patches, test_patches, val_labels, test_labels = train_test_split(
        temp_patches, temp_labels, test_size=0.5, random_state=42, stratify=temp_labels
    )

    print(f"\n--- Dataset Split Summary ---")
    print(f"Training patches: {len(train_patches)}, ({np.sum(train_labels == 1)} vessel, {np.sum(train_labels == 0)} non-vessel)")
    print(f"Validation patches: {len(val_patches)}, ({np.sum(val_labels == 1)} vessel, {np.sum(val_labels == 0)} non-vessel)")
    print(f"Test patches: {len(test_patches)}, ({np.sum(test_labels == 1)} vessel, {np.sum(test_labels == 0)} non-vessel)")

    # Calculate class weights for imbalance handling during training
    class_weights = compute_class_weight('balanced', classes=np.unique(train_labels), y=train_labels)
    class_weight_dict = {0: class_weights[0], 1: class_weights[1]}
    print(f"Class weights for training: {class_weight_dict}")

    # Data augmentation setup
    datagen = ImageDataGenerator(
        rotation_range=15,
        width_shift_range=0.1,
        height_shift_range=0.1,
        shear_range=0.05,
        zoom_range=0.05,
        horizontal_flip=True,
        vertical_flip=True,
        fill_mode='nearest'
    )

    # Build and compile the model
    model = build_model()
    # Define callbacks for learning rate reduction and early stopping
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=15, min_lr=0.00001, verbose=1)
    early_stopping = EarlyStopping(monitor='val_loss', patience=30, restore_best_weights=True, verbose=1)

    print("\nStarting model training...")
    history = model.fit(
        datagen.flow(train_patches, train_labels, batch_size=32),
        epochs=200, # Max epochs
        validation_data=(val_patches, val_labels),
        class_weight=class_weight_dict,
        callbacks=[reduce_lr, early_stopping],
        verbose=2 # Show one line per epoch
    )
    print("Model training finished.")

    # Evaluate model on validation set to find optimal threshold
    acc, auc, sensitivity, specificity, f1, iou, best_threshold = evaluate_model(model, val_patches, val_labels)
    if acc is not None:
        print(f"\nFinal Validation Metrics (Optimal Threshold: {best_threshold:.2f}):")
        print(f"Accuracy: {acc:.4f}, AUC: {auc:.4f}, Sensitivity: {sensitivity:.4f}, Specificity: {specificity:.4f}, F1-Score: {f1:.4f}, IoU: {iou:.4f}")

    # Plot training history
    plt.figure(figsize=(12, 4), dpi=100)
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.tight_layout()
    plt.savefig('history_plots.png')
    plt.show()

    # --- Evaluation on the held-out Test Set ---
    print("\n--- Evaluating model on the held-out Test Set ---")
    if len(test_patches) == 0:
        print("No test patches available for final evaluation.")
    else:
        test_loss, test_accuracy = model.evaluate(test_patches, test_labels, verbose=0)
        print(f"Test Loss: {test_loss:.4f}, Test Accuracy: {test_accuracy:.4f}")

        # Calculate full metrics on the test set using the best threshold found from validation
        test_predictions_probs = model.predict(test_patches, verbose=0).flatten() # Use probabilities for AUC
        test_pred_binary = (test_predictions_probs > (best_threshold if best_threshold is not None else 0.5)).astype(np.uint8)

        # Check for unique classes before calculating metrics that require them
        if len(np.unique(test_labels)) > 1 and len(np.unique(test_pred_binary)) > 1:
            test_auc = roc_auc_score(test_labels, test_predictions_probs) # Use probabilities for AUC
            test_f1 = f1_score(test_labels, test_pred_binary)
            test_iou = jaccard_score(test_labels, test_pred_binary)

            cm_test = confusion_matrix(test_labels, test_pred_binary, labels=[0, 1])
            tn, fp, fn, tp = cm_test.ravel()
            test_sensitivity = tp / (tp + fn) if (tp + fn) > 0 else 0
            test_specificity = tn / (tn + fp) if (tn + fp) > 0 else 0

            print(f"\nTest Set Metrics (Optimal Threshold: {best_threshold:.2f}):")
            print(f"Accuracy: {test_accuracy:.4f}, AUC: {test_auc:.4f}, Sensitivity: {test_sensitivity:.4f}, Specificity: {test_specificity:.4f}, F1-Score: {test_f1:.4f}, IoU: {test_iou:.4f}")

            # Plot Confusion Matrix for Test Set
            plt.figure(figsize=(8, 6), dpi=100)
            sns.heatmap(cm_test, annot=True, fmt='d', cmap='Blues', xticklabels=['Non-Vessel', 'Vessel'], yticklabels=['Non-Vessel', 'Vessel'])
            plt.title('Confusion Matrix on Test Patches')
            plt.xlabel('Predicted')
            plt.ylabel('True')
            plt.savefig('test_confusion_matrix.png')
            plt.show()

            print("\nClassification Report on Test Patches:")
            print(classification_report(test_labels, test_pred_binary, target_names=['Non-Vessel', 'Vessel']))
        else:
            print("Not enough unique classes in test labels or predictions to calculate full metrics (AUC, F1, IoU).")
            print(f"Unique test_labels: {np.unique(test_labels)}, Unique test_predictions: {np.unique(test_pred_binary)}")

    # --- Visualizing predictions on a few sample original test images ---
    # To visualize, we need to load original full images and apply prediction

    # Re-obtain the full list of paired files to select specific ones for visualization
    image_files_full_dataset = sorted(glob.glob(os.path.join(TEST_IMAGES_PATH, "*.ppm")))
    manual_files_full_dataset = sorted(glob.glob(os.path.join(TEST_MANUAL_PATH, "*.ppm")))

    paired_files_for_vis = []
    image_basenames_vis = {os.path.basename(f).replace('.ppm', ''): f for f in image_files_full_dataset}
    manual_basenames_vis = {}
    for f in manual_files_full_dataset:
        filename = os.path.basename(f)
        if '.ah.ppm' in filename:
            base_name_for_matching = filename.replace('.ah.ppm', '')
            manual_basenames_vis[base_name_for_matching] = f

    for basename in sorted(image_basenames_vis.keys()):
        if basename in manual_basenames_vis:
            paired_files_for_vis.append((image_basenames_vis[basename], manual_basenames_vis[basename]))

    # Select a few images for visualization from the complete set (e.g., first 5 or fewer if less available)
    num_vis_images = min(5, len(paired_files_for_vis))
    # It's challenging to guarantee these exact images contributed to `test_patches`
    # due to the random splitting of patches. For demonstration, we'll just pick the first few
    # images from the full dataset that are available.
    images_to_visualize = paired_files_for_vis[:num_vis_images]

    if not images_to_visualize:
        print("No images available for visualization of full mask predictions.")
    else:
        print(f"\n--- Generating and visualizing predictions for {len(images_to_visualize)} sample images ---")

        for i, (img_file, manual_file) in enumerate(images_to_visualize):
            print(f"Visualizing image {i + 1}/{len(images_to_visualize)}: {os.path.basename(img_file)}")

            original_image = load_image(img_file)
            ground_truth_manual = load_image(manual_file)

            if original_image is None or ground_truth_manual is None:
                print(f"Could not load image or manual for {os.path.basename(img_file)}. Skipping visualization.")
                continue

            ground_truth_manual_binary = (ground_truth_manual > 127).astype(np.uint8) * 255
            fov_mask = generate_fov_mask(original_image.shape)

            # Predict the mask for the full image using the optimal threshold
            predicted_mask = predict_mask(model, original_image, fov_mask, threshold=best_threshold if best_threshold is not None else 0.5)

            # Plotting the original, ground truth, and predicted masks
            plt.figure(figsize=(18, 6), dpi=100) # Increased figure size for better visibility
            plt.subplot(1, 3, 1)
            plt.imshow(original_image, cmap='gray')
            plt.title(f'Original Image\n{os.path.basename(img_file)}')
            plt.axis('off')

            plt.subplot(1, 3, 2)
            plt.imshow(ground_truth_manual_binary, cmap='gray')
            plt.title(f'Ground Truth ({np.sum(ground_truth_manual_binary > 0)} pixels)')
            plt.axis('off')

            plt.subplot(1, 3, 3)
            plt.imshow(predicted_mask, cmap='gray')
            plt.title(f'Predicted Mask ({np.sum(predicted_mask > 0)} pixels)')
            plt.axis('off')
            plt.tight_layout()
            viz_filename = f'sample_test_prediction_{os.path.basename(img_file).replace(".ppm", "")}.png'
            plt.savefig(viz_filename)
            plt.show()
            print(f"Saved individual visualization: {viz_filename}")

    # --- Final Output Check ---
    print("\nListing output directory contents:")
    print(os.listdir('/kaggle/working'))

    print("\nChecking and displaying generated plots and predictions:")
    output_files_to_check = ['history_plots.png', 'test_confusion_matrix.png']
    for file in output_files_to_check:
        if os.path.exists(f'/kaggle/working/{file}'):
            file_size = os.path.getsize(f'/kaggle/working/{file}') / 1024
            print(f"{file} exists, size: {file_size:.2f} KB")
            try:
                display(IPyImage(filename=f'/kaggle/working/{file}'))
            except Exception as e:
                print(f"Error displaying {file} via IPyImage: {e}")
                print(f"Please download /kaggle/working/{file} from Kaggle's output pane to view.")
        else:
            print(f"{file} not found!")

    # Display individual prediction images
    print("\nDisplaying individual image visualizations:")
    for i, (img_file, _) in enumerate(images_to_visualize):
        individual_filename = f'sample_test_prediction_{os.path.basename(img_file).replace(".ppm", "")}.png'
        if os.path.exists(individual_filename):
            print(f"Displaying {individual_filename}:")
            try:
                display(IPyImage(filename=individual_filename))
            except Exception as e:
                print(f"Error displaying {individual_filename} via IPyImage: {e}")
                print(f"Please download /kaggle/working/{individual_filename} from Kaggle's output pane.")
        else:
            print(f"{individual_filename} not found!")


if __name__ == "__main__":
    main()
```
